# Physics Room Results Analysis
Load event exports and produce quick plots/statistics.

```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt

events_path = Path('../physics engine/examples/events_sample.json')
events = json.loads(events_path.read_text()) if events_path.exists() else []
df = pd.DataFrame(events)
df.head()
```

```python
if not df.empty:
    h = df[df['event_type'] == 'hamiltonian.frame'].copy()
    h['energy'] = h['payload'].apply(lambda p: p.get('energy'))
    h.plot(x='sequence', y='energy', title='Energy vs Sequence')
    plt.show()
else:
    print('No events loaded. Export events to physics engine/examples/events_sample.json')
```

```python
if not df.empty:
    s = df[df['event_type'] == 'stability.harmonic'].copy()
    s['drift'] = s['payload'].apply(lambda p: p.get('drift'))
    s['q_factor'] = s['payload'].apply(lambda p: p.get('q_factor'))
    print(s[['drift', 'q_factor']].describe())
    s.plot(x='sequence', y=['drift', 'q_factor'], title='Harmonic Stability Metrics')
    plt.show()
else:
    print('No stability frames available.')
```
